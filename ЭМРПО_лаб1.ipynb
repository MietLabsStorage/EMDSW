```python
import math 
import matplotlib.pyplot as plt

times = [1, 60, 60*60, 60*60*24, 60*60*24*30, 60*60*24*365, 60*60*24*365*100]

def lg_n(t):
  return 10**t


def sqrt_n(t):
  return t**2


def n(t):
  return t


def n_lg_n(t):
  n = 1;
  while(n*math.log10(n) < t):    
    n+=1    
  return n  


def n_2(t):
  return math.sqrt(t)
  

def n_3(t):
  return t**(1/3)
  

def power2_n(t):
  return math.log2(t)
  

def factorial_n(t):
  n = 1;
  while(math.factorial(n) < t):    
    n+=1    
  return n  


def plot(title, X, Y):
  plt.title(title)
  plt.xlabel('n')
  plt.ylabel('t')
  plt.grid()
  plt.plot(X, Y)
```

### lg(n):

```python
T = times[:4]
N = [lg_n(t) for t in T]
print([str(n)+' ' for n in N])
# plot('lg(n)', N, T)
```

### sqrt(n):

```python
T = times[:7]
N = [sqrt_n(t) for t in T]
print([str(n)+' ' for n in N])
plot('sqrt(n)', N, T)
```

### n:

```python
T = times[:7]
N = [n(t) for t in T]
print([str(n)+' ' for n in N])
plot('n', N, T)
```

### n*lg(n):

```python
T = times[:6]
N = [n_lg_n(t) for t in T]
print([str(n)+' ' for n in N])
plot('n*lg(n)', N, T)
```

### n^2(n):

```python
T = times[:7]
N = [n_2(t) for t in T]
print([str(n)+' ' for n in N])
plot('n^2', N, T)
```

### n^3(n):

```python
T = times[:7]
N = [n_3(t) for t in T]
print([str(n)+' ' for n in N])
plot('n^3', N, T)
```

### 2^n(n):

```python
T = times[:7]
N = [power2_n(t) for t in T]
print([str(n)+' ' for n in N])
plot('2^n', N, T)
```

### n!(n):

```python
T = times[:7]
N = [factorial_n(t) for t in T]
print([str(n)+' ' for n in N])
plot('n!', N, T)
```
